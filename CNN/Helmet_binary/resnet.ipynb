```python
# base_path : 이미지가 있는 경로
# label_path : 좌표값 txt가 있는 경로
# save_path : crop 이미지 저장할 경로

def image_crop(image_path, label_path,save_path):
    label = glob(label_path + "*.txt")

    for i in tqdm(range(len(label))):
        with open(label[i]) as f:
            # .txt를 제거한 파일 이름만 추출
            #  ex) f993aefbee7c9b59
            img_name = label[i].split('Label/')[1].split('.txt')[0]

            # 추출한 이름에 .jpg
            img_name = img_name +'.jpg'

            # 이미지 경로 + 이미지 이름
            img = cv2.imread(image_path + img_name)

            # 한 이미지당 2개 이상의 좌표시 번호 붙이기
            cnt = 0
            for line in f.read().splitlines():
                # 텍스트 폴더에서 클래스 이름뺴고 좌표만 추출, 뒤에서 4개
                # 좌표값이 문자열형태임 -> 실수형
                x ,y , xw, yh = list(map(float,line.split()[-4:]))

                # path : crop 이미지 저장폴더
                path = save_path

                # 저장
                cv2.imwrite(path + f'{str(cnt)}{img_name}',img[int(y):int(yh),int(x): int(xw)])
                # 중복안되게 cnt 증가
                cnt +=1
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D
from keras.layers import Dropout
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
import tensorflow as tf
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.models import Model
```

```python
model = ResNet50(weights='imagenet')

x = model.output
x = Flatten()(x)
x = Dense(512, activation='relu')(x)
x = Dropout(0.5)(x)
x = Dense(256, activation='relu')(x)

# 예측 : 헬멧썻?안썼?->2 
predictions = Dense(1, activation='sigmoid')(x)


model = Model(inputs=model.input, outputs=predictions)

model.summary()
```

## Model compile() 
- 모델을 학습시키기 이전에, compile() 메소드를 통해서 학습 방식에 대한 환경설정을 해야 합니다.

- 다음 세 개의 인자를 입력으로 받습니다.

- 정규화기 (optimizer)

  - 훈련과정을 설정합니다. 즉, 최적화 알고리즘을 설정을 의미합니다.
  - adam, sgd, rmsprop, adagrad 등이 있습니다.

- 손실 함수 (loss function)

  - 모델이 최적화에 사용되는 목적 함수입니다.
  - mse, categorical_crossentropy, binary_crossentropy 등이 있습니다.

- 평가지표 (metric)

  - 훈련을 모니터링 하기 위해 사용됩니다.
  - 분류에서는 accuracy, 회귀에서는 mse, rmse, r2, mae, mspe, mape, msle 등이 있습니다.
  - 사용자가 메트릭을 정의해서 사용할 수도 있습니다.
  - metrics 관련 정리 [https://wooono.tistory.com/99]

```python
model.compile(    
                  # 훈련과정을 설정합니다. 즉, 최적화 알고리즘을 설정
                 optimizer=tf.keras.optimizers.Adam(1e-5) 
                 # 모델이 최적화에 사용되는 목적 함수
              # 3개이상 분류할 때 :categorical_crossentropy
                 ,loss='binary_crossentropy'
                  # 정확도 몇인지 확인
                 ,metrics=['accuracy']
             )
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import numpy as np
```

```python
train_path = '/content/drive/MyDrive/cut_data_limit10000/train/'
test_path = '/content/drive/MyDrive/cut_data_limit10000/test/'
vali_path = '/content/drive/MyDrive/cut_data_limit10000/validation/'
```

```python
# 데이터셋 이미지 로드

# 증강하지 않고 옵션 한개만 줌
train_datagen = ImageDataGenerator(
    rescale = 1/255 # 255로 이미지 나눔
    )


train_generator = train_datagen.flow_from_directory(
    train_path, # crop/yes or no
    target_size = (224,224), # resnet 이 input size 224로 받는것 같음
    batch_size = 1,# 한장씩 받을것
    class_mode = 'binary', # 이진분류
    shuffle = True
)

# 테스트셋
test_datagen = ImageDataGenerator(
    rescale = 1/255
    )
test_generator = test_datagen.flow_from_directory(
    test_path,
    target_size = (224,224),
    
    batch_size = 1,
    class_mode = 'binary',
    shuffle = False
)
```

```python
vali_datagen = ImageDataGenerator(
    rescale = 1/255
    )
vali_generator = vali_datagen.flow_from_directory(
    vali_path,
    target_size = (224,224),
    batch_size = 1,
    class_mode = 'binary',
    shuffle = False
)
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint
```

```python
%cd /content
!mkdir check

# checkpoint 만들기
checkpoint_path = '/content/check/tmp_checkpoint.ckpt'

checkpoint = ModelCheckpoint(
    filepath=checkpoint_path
    ,save_weights_only=True
    ,save_best_only=True
    ,monitor='loss' # val_loss
    ,verbose=1
)
```

```python
model.fit(
	train_generator 
	,validation_data = vali_generator
 	,batch_size=200
	,steps_per_epoch=5 # lr : 1e-4 / steps_per_epoch : train_generator 에서 얼마나 많은 샘플을 뽑을것인가 
	,epochs=500
	,callbacks=[checkpoint]
)
```

```python
#evaluate 모델 평가 -> test셋으로 해야함
model.evaluate(
    test_generator #(x_test, y_test)
    , batch_size = 8 
)
```
