```python
#증강 이미지를 저장할 위치 선택
save_here = '/content/drive/MyDrive/final_project_team8/킥보드 탑승 예시사진/make_yes'
```

```python
# 이미지를 증강하는 ImageGenerator 객체 생성
datagen = ImageDataGenerator(
	rotation_range=45, # 이미지 회전 각도
	zoom_range=0.15, # 이미지 확대 비율
	width_shift_range=0.2, # 이미지 가로 이동 비율  
	height_shift_range=0.2, # 이미지 세로 이동 비율
	shear_range=0.15, # 이미지 기울기 비율
	horizontal_flip=True,# 이미지 좌우 반전
	fill_mode="nearest", # 이미지를 회전 이동시 빈공간을 채우는 방식
    rescale=1/255 # 이미지에 1/255 를 곱해줌
	)
```

```python
len(test)
```

```python
for i in range(len(test)):
  image_path = test[i]
  image = np.expand_dims(ndimage.imread(image_path), 0)

  #원본 이미지를 맞춥니다.
  datagen.fit(image)
  
  #이미지를 반복하고 "save_to_dir" 매개변수를 사용하여 저장
  for x, val in zip(datagen.flow(image,                    #image we chose
          save_to_dir=save_here,     #this is where we figure out where to save
          save_prefix='aug',        # it will save the images as 'aug_0912' some number for every new augmented image
          save_format='jpg'),range(4)) :pass     # here we define a range because we want 10 augmented images otherwise it will keep looping forever I think
```
